```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings

#warnings.filterwarnings('ignore')
```

Сперва загружаем данные

```python
mpgData = sns.load_dataset('mpg')
carsData = pd.read_csv('CarsData.csv')
obsData = pd.read_csv('obesity.csv')
```

Это график рассеивания, где сравниваются год выпуска машин и их вес. </br>
Также здесь есть rugplot, который является дополнением к графику рассеивания и показывает метки значений на оси OX.

```python
plt.xlabel('Вес(кг)')
plt.ylabel('Год выпуска')
plt.title('Сравнение веса машин в зависимости от года выпуска')
sns.scatterplot(data=mpgData, x=mpgData['weight'], y=mpgData['model_year'])
sns.rugplot(data=mpgData, x=mpgData['weight'], y=mpgData['model_year'])
```

Здесь также используется график рассеивания для сравнения стоимости машин в зависимости от пробега.

```python
plt.xlabel('Пробег в милях')
plt.ylabel('Цена в евро')
plt.title('Сравнение стоимости машин в зависимости от пробега')
sns.scatterplot(data=carsData, x=carsData['mileage'], y=carsData['price'])
```

Чтобы отличить тип трансмиссии у автомобилей добавим свойство **hue=carsData['transmission'].** </br>
Это свойство разделяет данные на группы по цветам.
Теперь будут заметны какие машины дороже в зависимости от трансмиссии и пробега.

```python
plt.xlabel('Пробег в милях')
plt.ylabel('Цена в евро')
plt.title('Сравнение стоимости машин в зависимости от пробега')
sns.scatterplot(data=carsData, x=carsData['mileage'], y=carsData['price'], hue=carsData['transmission'])
```

Здесь используется график jointplot с теми же данными. Добавлен параметр **kind='hex'**. </br>
Как я понял, он показывает плотность данных в определенном месте на графике.

```python
sns.jointplot(data=carsData, x=carsData['mileage'], y=carsData['price'], kind='hex') 
sns.rugplot(data=carsData, x=carsData['mileage'], y=carsData['price']) # дополнительно показывает отдельные значения
plt.xlabel('Пробег в милях')
plt.ylabel('Цена в евро')
plt.title('Стоимость машин в зависимости от пробега')
```

Согласно документации, график kdeplot нужен для построения данных используя ядерную оценку плотности. </br>
Параметр **multiple** отвечает за метод отрисовки данных.

```python
sns.kdeplot(data=carsData, x='year', hue='transmission', multiple='stack')
plt.title('Зависимость года производства и типа трансмиссии')
plt.xlabel('Год')
plt.ylabel('Оценка плотности') # density
```

Здесь используются те же данные, но изменен метод отрисовки на **заполнение "fill"**

```python
sns.kdeplot(data=carsData, x='year', hue='transmission', multiple='fill')
plt.title('Зависимость года производства и типа трансмиссии')
plt.xlabel('Год')
plt.ylabel('Оценка плотности') # density
```

```python
sns.kdeplot(data=carsData, x='year', hue='transmission', multiple='layer')
plt.title('Зависимость года производства и типа трансмиссии')
plt.xlabel('Год')
plt.ylabel('Оценка плотности') # density
```

График regplot сравнивает данные и добавляет линию регрессии.

```python
sns.regplot(data=carsData, x=carsData['mileage'], y=carsData['price'])
plt.xlabel('Пробег в милях')
plt.ylabel('Цена в евро')
plt.title('Сравнение цены в зависимости от пробега')
```

Здесь используется график violinplot для сравнения расхода топлива в зависимости от количества цилиндров в двигателе.

```python
plt.title('Сравнение расхода топлива в зависимости от количества цилиндров')
sns.violinplot(data=mpgData, x='cylinders', y='mpg')
plt.xlabel('Количество цилиндров')
plt.ylabel('Расход топлива (mpg)')
```

На графике displot показано какой налог имеет большинство автомобилей в данных. </br>
Параметр **bins=15** делит данные на интервалы и только потом строит гистограмму. 

```python
sns.displot(data=carsData, x='tax', bins=15)
plt.xlabel('Налоговая ставка(tax rate)')
plt.ylabel('Количество машин')
plt.title('Сравнение налоговой ставки по отношению к автомобилям')
```

Здесь показана стоимость машин, которые присутствуют в датасете. </br>
Также на гистограмму добавлена KDE кривая **kde=True**

```python
import matplotlib.ticker as ticker

sns.displot(data=carsData, x='price', kde=True, bins=20) # с добавлением кривой
plt.title('График для оценки стоимости машин')
plt.xlabel('Стоимость в евро')
plt.xticks(range(0, 170000, 25000))
plt.ylabel('Количество машин')
```

Здесь используется другой вид графика **kind="kde"** для данных о расходе топлива. 

```python
sns.displot(data=mpgData, x='mpg', kind='kde')
plt.title('Ядерная оценка плотности для расхода топлива')
plt.xlabel('Расход топлива(mpg)')
plt.ylabel('Оценка плотности')
```

Здесь создается сетка и используется диаграмма histplot для каждого вида топлива.

```python
group = sns.FacetGrid(carsData, col='fuelType')
group.map_dataframe(sns.histplot, x='price')
group.set_axis_labels(x_var='Цена в евро', y_var='Количество авто')
```

Здесь используется график boxplot для сравнения расхода топлива машин от года производства

```python
sns.boxplot(data=carsData, x=carsData['mpg'],y=carsData['year'], orient='h')
plt.title('Сравнение расхода топлива машин от года производства')
plt.xlabel('Расход топлива(mpg)')
plt.ylabel('Год')
```

Здесь используется такой же график, но для другого набора данных. </br>
С этими данными он получился получше.

```python
sns.boxplot(data=mpgData, x=mpgData['mpg'],y=mpgData['model_year'], orient='h')
plt.title('Оценка расхода топлива по годам')
plt.xlabel('Расход топлива(mpg)')
plt.ylabel('Год')
```

В графике stripplot я отобразил как влияет объем двигателя на цену автомобиля.

```python
sns.stripplot(data=carsData, x='engineSize', y='price')
plt.title('Стоимость авто в зависимости от объема двигателя')
plt.xlabel('Объем двигателя')
plt.xticks(rotation=90)
plt.ylabel('Цена в евро')
```

График pointplot состоит из соединяющих линий и точек. </br>
В графике ниже также идет разбиение по типу топлива.

```python
sns.pointplot(data=carsData, x='transmission', y='price', hue='fuelType')
plt.title('График зависимости цены от типа трансмиссии')
plt.xlabel('Трансмиссия')
plt.ylabel('Цена')
```

В графике ниже показана зависимость расхода топлива от количества лошадиных сил с линией регрессии.

```python
sns.lmplot(data=mpgData, x='horsepower', y='mpg')
plt.title('Зависимость расхода топлива от количества лошадиных сил')
plt.xlabel('Лошадиные силы')
plt.ylabel('Расход топлива(mpg)')
```

Здесь представлена обычная линейная диаграмма, построенная с помощью matplotlib. </br>
На графике показано ожирение России в процентах с 1975 по 2016.

```python
russiaData = obsData[obsData['Entity'] == 'Russia']
plt.plot(russiaData['Year'], russiaData['Indicator:Prevalence of obesity among adults, BMI &GreaterEqual; 30 (crude estimate) (%) - Sex:Both sexes'],
           linewidth=2,
           label='Россия',
           color='blue')
plt.title('Ожирение в России')
plt.xlabel('Год')
plt.ylabel('Ожирение (%)')
plt.grid()
plt.legend()
```

Здесь представлены такие же данные, но в виде столбчатой диаграммы.

```python
import matplotlib.ticker as ticker

sns.barplot(data=russiaData, x=russiaData['Year'], y=russiaData['Indicator:Prevalence of obesity among adults, BMI &GreaterEqual; 30 (crude estimate) (%) - Sex:Both sexes'])
plt.title('Ожирение в России')
plt.xlabel('Год')
plt.xticks(range(0, len(russiaData)+1, 5))
plt.ylabel('Ожирение (%)')
plt.grid()
```
